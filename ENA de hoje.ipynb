```python
import pandas as pd
import matplotlib.pyplot as plt
from src import trata_acomph, regressoes, ena
```

```python
ena_g = ena.calc_ena()
```

# ENA por submercado

```python
ena_s = ena.ena_mercados(ena_g)
ax = ena_s.T.iloc[20:30,:].plot(kind = 'bar', subplots = True, figsize = [15, 9])
for j in range(4):
    leg = ax[j].get_legend().remove()
    for i in ax[j].patches:
        ax[j].text(i.get_x()+.07, i.get_height()-i.get_height()*.3, \
            str(round(i.get_height()/1000,2)), fontsize=11,
                color='black')
    
plt.show()
```
